```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_maps(maps):
    zmax = 1000.0
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(5, 8), dpi=600)
    
    im1 = ax1.imshow(maps[0].T,vmin=0,vmax=zmax,origin='lower',cmap='jet')
    ax1.set_title("Plane 0")
    ax1.set_xlabel("Wire")
    ax1.set_ylabel("Time Tick")

    im2 = ax2.imshow(maps[1].T,vmin=0,vmax=zmax,origin='lower',cmap='jet')
    ax2.set_title("Plane 1")
    ax2.set_xlabel("Wire")

    im3 = ax3.imshow(maps[2].T,vmin=0,vmax=zmax,origin='lower',cmap='jet')
    ax3.set_title("Plane 2")
    ax3.set_xlabel("Wire")

    plt.tight_layout()
```

```python
plot_maps(X[1])
```
